# Bi Variate Analysis

```python
sns.scatterplot(data = df, x = 'number_of_reviews', y = 'price')
plt.title('Price vs Number of Reviews')
plt.xlabel('Number of Reviews')
plt.ylabel('Price')
plt.show()
```

**Observation :**

*As the number of reviews increase the price is decreases they both are in inversely proportional relationship*

```python
df.columns
```

```python
df.groupby('room_type')['number_of_reviews'].max().reset_index()
```

**Observation :**

- *As we see the Hotel Room has the max number of reviews*

```python
corr = df[['price', 'minimum_nights', 'number_of_reviews', 'reviews_per_month', 'calculated_host_listings_count', 'availability_365','beds']].corr()
corr
```

```python
sns.heatmap(data =corr, annot = True, cmap = 'coolwarm' )
plt.title('Correlation Heatmap')
plt.figure(figsize=(6,6))
plt.show()
```

**Observation :**

- *As the price increases the number of beds also increases*
